# Atividade 0
### Rotação horária de 90° e shear (cisalhamento) em x

#### Vetor Inicial

### Vetor Inicial:
O vetor escolhido foi o seguinte:
\begin{bmatrix}
4 \\
6
\end{bmatrix}

```python

import matplotlib.pyplot as plt

plt.quiver(0,0,4,6,scale=1,scale_units='xy',angles='xy')
plt.xlim(0,12)
plt.ylim(0,12)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
```

### Vetor Após Rotação:

Após ter ume vetor arbitrário escolhido foi realizada a rotação do mesmo em 90° em sentido horário, assim se obtendo o seguinte vetor:

\begin{bmatrix}
4 \\
-6
\end{bmatrix}

```python
import matplotlib.pyplot as plt

plt.quiver(0,0,4,-6,scale=1,scale_units='xy',angles='xy')
plt.xlim(-3,4)
plt.ylim(-6,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
```

### Vetor Após Rotação e Cisalhamento:

Com a rotação já realizada, fazemos então o cisalhamento que basicamente resultou no deslocamento para a esquerda do vetor, resultando no seguinte vetor:

\begin{bmatrix}
-2 \\
-6
\end{bmatrix}

```python
import matplotlib.pyplot as plt

plt.quiver(0,0,-2,-6,scale=1,scale_units='xy',angles='xy')
plt.xlim(-6,3)
plt.ylim(-6,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
```

# Obtenção das matrizes de transformação

#### Explicação 
Primeiro são definidos os vetores unitários que são a base do sistema com valor unitário igual a 1

```python

import matplotlib.pyplot as plt

plt.quiver(0,0,0,1,scale=1,scale_units='xy',angles='xy')
plt.quiver(0,0,1,0,scale=1,scale_units='xy',angles='xy')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.text(-0.20,0.40,'j')
plt.text(0.40,-0.30,'i')
plt.gca().set_aspect('equal', adjustable='box')
```

### Vetor j:

\begin{bmatrix}
0 \\
1
\end{bmatrix}

### Vetor i:

\begin{bmatrix}
1 \\
0
\end{bmatrix}

### Matriz resultante
Após a concatenação:
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}


### Após a Rotação


```python

import matplotlib.pyplot as plt

plt.quiver(0,0,1,0,scale=1,scale_units='xy',angles='xy')
plt.quiver(0,0,0,-1,scale=1,scale_units='xy',angles='xy')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.text(0.40,0.20,'j')
plt.text(-0.25,-0.5,'i')
plt.gca().set_aspect('equal', adjustable='box')
```

### Explicação 
Após a rotação de 90° em sentido horário dos vetores que formam a base, os valores dos vetores passaram a ser os seguintes:

### Vetor j:

\begin{bmatrix}
1 \\
0
\end{bmatrix}

### Vetor i:

\begin{bmatrix}
0 \\
-1
\end{bmatrix}

### Matriz resultante da Rotação
Após a concatenação:
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}

### Após o Cisalhamento


```python
import matplotlib.pyplot as plt

plt.quiver(0,0,1,1,scale=1,scale_units='xy',angles='xy')
plt.quiver(0,0,1,0,scale=1,scale_units='xy',angles='xy')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.text(0.4,0.6,'j')
plt.text(0.40,-0.30,'i')
plt.gca().set_aspect('equal', adjustable='box')
```

### Explicação
Após o Cisalhamento o vetor j é inclinado para a direita e o vetor i se mantem o mesmo, com os dois ficando com os seguintes valores:


### Vetor j:

\begin{bmatrix}
1 \\
1
\end{bmatrix}

### Vetor i:

\begin{bmatrix}
1 \\
0
\end{bmatrix}

### Matriz resultante do Cisalhamento
Após a concatenação:
\begin{bmatrix}
1 & 1 \\
1 & 0
\end{bmatrix}

### Após a Rotação e o Cisalhamento

```python

import matplotlib.pyplot as plt

plt.quiver(0,0,1,0,scale=1,scale_units='xy',angles='xy')
plt.quiver(0,0,-1,-1,scale=1,scale_units='xy',angles='xy')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.axvline(0,color='black',linewidth=1)
plt.text(0.40,0.20,'j',fontweight='bold')
plt.text(-0.55,-0.3,'i',fontweight='bold')
plt.gca().set_aspect('equal', adjustable='box')
```

### Explicação 
Após a rotação e o Cisalhamento, ocorreu um deslocamento do vetor i para a esquerda e para baixo, o que fez o valor dos vetores ficaram iguais a:

### Vetor j:

\begin{bmatrix}
1 \\
0
\end{bmatrix}

### Vetor i:

\begin{bmatrix}
-1 \\
-1
\end{bmatrix}


#### Obtenção da Matriz Transformação
A seguir é feito a concatenação dos dois vetores e então é criada uma matriz transformação:

\begin{bmatrix}
1 & -1 \\
0 & -1
\end{bmatrix}
